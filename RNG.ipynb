```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_uniform(a, num_samples):
    samples = np.random.rand(num_samples) * 2 * a - a
    return samples

def visualize_histogram(data, bins):
    plt.hist(data, bins=bins, density=True, alpha=0.7, color='blue')
    plt.title('Histogram of Random Numbers')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

# Parameters
a = 10  # Range
num_samples = 10000  # Number of samples
bins = 50  # Number of bins for histogram

# Generate random numbers using Monte Carlo method
random_numbers = monte_carlo_uniform(a, num_samples)

# Visualize the results using a histogram
visualize_histogram(random_numbers, bins)
```
