```python
import numpy as np
import matplotlib.pyplot as plt

def generate_uniform_random_numbers(n, a):
    """
    Generates n random numbers from the range [-a, a] with a uniform distribution.
    """
    return np.random.uniform(low=-a, high=a, size=n)

def visualize_histogram(data, a):
    """
    Visualizes the histogram of the generated random numbers.
    """
    plt.hist(data, bins=50, density=True, alpha=0.6, color='g')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Uniform Random Numbers from [-{a}, {a}]')
    plt.grid(True)
    plt.show()

def main():
    # Parameters
    n = 10000  # Number of random numbers to generate
    a = 10     # Range parameter

    # Generate random numbers
    random_numbers = generate_uniform_random_numbers(n, a)

    # Visualize the histogram
    visualize_histogram(random_numbers, a)

if __name__ == "__main__":
    main()
```
